```python
loan = pd.get_dummies(loan, columns=["person_education"], dtype=np.int8)
loan
```

```python
loan = pd.get_dummies(loan, columns=["person_home_ownership"], dtype=np.int8)
loan
```

```python
loan = pd.get_dummies(loan, columns=["loan_intent"], dtype=np.int8)
loan
```

```python
loan["previous_loan_defaults_on_file"] = [1 if i == "Yes" else 0 for i in loan["previous_loan_defaults_on_file"]]
loan
```

```python
loan.loc[:, "person_income"] = np.log(loan["person_income"])
loan
```

```python
loan.loc[:, "loan_amnt"] = np.log(loan["loan_amnt"])
loan
```

```python
loan.columns
```

```python
X = loan[['person_age', 'person_gender', 'person_income', 'person_emp_exp',
       'loan_amnt', 'loan_int_rate', 'loan_percent_income',
       'cb_person_cred_hist_length', 'credit_score',
       'previous_loan_defaults_on_file',
       'person_education_Associate', 'person_education_Bachelor',
       'person_education_Doctorate', 'person_education_High School',
       'person_education_Master', 'person_home_ownership_MORTGAGE',
       'person_home_ownership_OTHER', 'person_home_ownership_OWN',
       'person_home_ownership_RENT', 'loan_intent_DEBTCONSOLIDATION',
       'loan_intent_EDUCATION', 'loan_intent_HOMEIMPROVEMENT',
       'loan_intent_MEDICAL', 'loan_intent_PERSONAL', 'loan_intent_VENTURE']]

X
```

```python
y = loan['loan_status']
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.3)
```

```python
logisticRegression = LogisticRegression(max_iter=5000)
logisticRegression
```

```python
logisticRegression.fit(X_train, y_train)
logisticRegression
```

```python
logisticRegression.coef_
```

```python
logisticRegression.intercept_
```

```python
y_predict = logisticRegression.predict(X_test)
print(y_predict)
# print(y)
```

```python
cm = confusion_matrix(y_test, y_predict)
cm
```
